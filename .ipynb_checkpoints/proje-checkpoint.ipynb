```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
main_url = "https://www.klinikbewertungen.de/klinik-forum/erfahrung-mit-krankenhaus-soltau"
req = requests.get(main_url)
req.encoding='utf-8'
soup = BeautifulSoup(req.text, "html.parser")
```

# Voll texte Webscrapping

```python
count=1
for a in soup.find_all('article',class_ = "bewertung"):
    bewertung = a.text.rstrip().lstrip()
    print('\n+++++++++++++++++++' ,(count), 'REVIEW +++++++++++++++++++')
    print(bewertung)
    count+=1
```

# Klinik name

```python
klinik_name = soup.find('h1').text
klinik_name
```

# Alle Titel

```python
Titels = soup.find_all('h2')
for t in Titels:
    print(t.text)
```

# Alle Bewertung DATUM

```python
Datum = soup.find_all('time')
for dat in Datum:
    print(dat.text)
```
