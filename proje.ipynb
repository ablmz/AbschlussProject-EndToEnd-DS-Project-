```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
main_url = "https://www.klinikbewertungen.de/klinik-forum/erfahrung-mit-krankenhaus-soltau"
req = requests.get(main_url)
soup = BeautifulSoup(req.text, "html.parser")
```

```python
comments = soup.find_all("article", class_ = "bewertung")
```

```python
for comment in comments:
    print(comment)
    print("---------------------------------")
```

```python
titles = soup.find_all("h2")
```

```python
titles
```

```python
for title in titles:
    print(title.text)
```

```python
commentTexts=soup.find_all("p")
```

```python
for comment in commentTexts:
    print(comment)
    print("\n")
```
