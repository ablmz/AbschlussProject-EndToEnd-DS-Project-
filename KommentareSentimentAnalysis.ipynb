```python
import pandas as pd
import math
```

```python
import textblob
from textblob_de import TextBlobDE
```

```python
reviews = pd.read_csv("output.csv")
```

```python
gData=pd.read_csv("ggl_scrap_allurl_22.csv")
```

```python
gData["Textuelle_Bewertung"]
```

```python
sentiment=[]
sentimentBewertung=[]
for komment in gData["Textuelle_Bewertung"]:
    if "nan" in str(komment):
        print("")
        sentiment.append("")
        sentimentBewertung.append("")
    else:
        print(komment)
        blob = TextBlobDE(str(komment))
        ergebnis=blob.sentiment
        sentiment.append(ergebnis)
        print(ergebnis)
        print("------------------------")
        if(ergebnis[0]>0):
            sentimentBewertung.append("Positiv")
            print("Positiv")
        elif(ergebnis[0]<0):
            sentimentBewertung.append("Negativ")
            print("Negativ")
        else:
            sentimentBewertung.append("Neutral")
            print("Neutral")
        
            
        
        
        
        
```

```python
gData['sentiment']=sentiment
```

```python
gData['sentimentBewertung']=sentimentBewertung
```
